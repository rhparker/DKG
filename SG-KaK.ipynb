```python
# import os module
import os

# import AUTO modules
from auto import AUTOCommands as ac
from auto import AUTOclui as acl
from auto import interactiveBindings as ib
from auto import runAUTO as ra

import numpy as np

# import plotting tools
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec

# virtual display to use AUTO plot function
from pyvirtualdisplay import Display
from IPython.display import display as display
from PIL import Image
disp = Display(visible=0, size=(1366,768))
disp.start()
```

```python
# start AUTO with runner object
runner = ra.runAUTO()

r1 = ac.run(e='ab', c='ab', UZR={1:0.25}, runner=runner)
```

```python
r1 = ac.loadbd()

# these are the types and labels for algebraic problems
# "user output point" is type number -4, so this works for that
AUTOtypes = ["","BP","LP","HB","","","UZ","","","EP"]
AUTOlabels = ["","branch point","fold","Hopf bifurcation","regular output point","","user output point","",
                  "", "end point"]
NDim = r1[0].c['NDIM']
# branches of the solution
branches = [ 
         { "par"       : br.coordarray[0,:], 
           "L2norm"    : br.coordarray[1,:],
           "branchnum" : br.BR
         } for br in r1 ]

# individual solutions
# sols = [ {  "u"         : np.array( s[0]['u'] ),
#             "L2norm"    : np.linalg.norm( s[0]['u'] ),
#             "par"       : s.PAR[0],
#             "branchnum" : s.data['BR'],
#             "typenum"   : s.data['TY number'],
#             "type"      : AUTOtypes[ s.data['TY number'] ],
#             "typename"  : AUTOlabels[ s.data['TY number'] ]
#          } for s in r1()  ]

# solutions of interest
sUZ = [ {   "u"         : np.array( s[0]['u'] ),
            "L2norm"    : np.linalg.norm( s[0]['u'] ),
            "par"       : s.PAR[0],
            "branchnum" : s.data['BR'],
            "typenum"   : s.data['TY number'],
            "type"      : AUTOtypes[ s.data['TY number'] ],
            "typename"  : AUTOlabels[ s.data['TY number'] ]
         } for s in r1() if AUTOtypes[ s.data['TY number'] ] == 'UZ' ]

sBP = [ {   "u"         : np.array( s[0]['u'] ),
            "L2norm"    : np.linalg.norm( s[0]['u'] ),
            "par"       : s.PAR[0],
            "branchnum" : s.data['BR'],
            "typenum"   : s.data['TY number'],
            "type"      : AUTOtypes[ s.data['TY number'] ],
            "typename"  : AUTOlabels[ s.data['TY number'] ]
         } for s in r1() if AUTOtypes[ s.data['TY number'] ] == 'BP' ]

sLP = [ {   "u"         : np.array( s[0]['u'] ),
            "L2norm"    : np.linalg.norm( s[0]['u'] ),
            "par"       : s.PAR[0],
            "branchnum" : s.data['BR'],
            "typenum"   : s.data['TY number'],
            "type"      : AUTOtypes[ s.data['TY number'] ],
            "typename"  : AUTOlabels[ s.data['TY number'] ]
         } for s in r1() if AUTOtypes[ s.data['TY number'] ] == 'LP' ]

othersol = [120]
others = [ r1(ind) for ind in othersol ]

sother = [ {   "u"      : np.array( s[0]['u'] ),
            "L2norm"    : np.linalg.norm( s[0]['u'] ),
            "par"       : s.PAR[0],
            "branchnum" : s.data['BR'],
            "typenum"   : s.data['TY number'],
            "type"      : AUTOtypes[ s.data['TY number'] ],
            "typename"  : AUTOlabels[ s.data['TY number'] ]
         } for s in others ]
```

```python
# plot bifurcation diagram
fontsize=16
fig = plt.figure()

# whole thing
for branch in branches:
    bcolor = 'C'+str( branch["branchnum"] - 1 )
    plt.plot( branch['par'], branch['L2norm'], c=bcolor, linewidth=2)
    plt.xlabel('$d$',fontsize=fontsize);
    plt.ylabel('$||u_n||_2$',fontsize=fontsize);    
    plt.xlim(1.19,1.206)
    plt.ylim(20.5175,20.5325)
    
plt.savefig('kak8.eps')
```

```python
# spectral stuff

n = NDim
D2 = np.diag( -2*np.ones(n) ) + np.diag( np.ones(n-1),1 ) + np.diag( np.ones(n-1),-1 )
D2[0,1] = 2;
D2[-1,-2] = 2;

# ind = 1
# u = sUZ[ind]['u'];
# d = sUZ[ind]['par']

ind = 0
u = sLP[ind]['u'];
d = sLP[ind]['par']

ind = 0
u = sother[ind]['u'];
d = sother[ind]['par']

J = d*D2 + np.diag( np.cos(u) )
l = np.linalg.eigvals(J)
l2 = np.lib.scimath.sqrt(l)

plt.figure()
plt.plot( np.real(l2), np.imag(l2), 'o' )
plt.plot( -np.real(l2), -np.imag(l2), 'o' )
```
